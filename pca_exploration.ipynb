```python
%matplotlib inline
from os import listdir, path
import numpy as np
from matplotlib import pyplot as plt
from sklearn.decomposition import PCA

import pickle

plt.rcParams["figure.figsize"] = (20,10)

data_path = 'datasets/100x100'
```

```python
with open(path.join(data_path, 'images'), 'rb') as f:
    images = pickle.load(f)

with open(path.join(data_path, 'labels'), 'rb') as f:
    labels = pickle.load(f)
    
print(labels[0, 4:])
print(images[0])

print(len(images), len(labels))
print('Input shape: ', images.shape)
print('Input dimension: ', images[0].shape[0] * images[0].shape[1] * images[0].shape[2])
```

```python
for i in range(10):
    print(labels[i])
    plt.imshow(images[i])
    plt.show()
```

```python
flattened_images = images.reshape((images.shape[0], -1))
```

```python
pca = PCA(n_components=500)
pca.fit(flattened_images)

plt.plot(np.cumsum(pca.explained_variance_ratio_))
plt.xlabel('Number of components')
plt.ylabel('Cumulative explained variance')
```

```python
NCOMPONENTS = 500

pca = PCA(n_components=NCOMPONENTS)
pca_images = pca.fit_transform(flattened_images)
pca_std = np.std(pca_images)

print(pca_images.shape)
```

```python
inv_pca_images = pca.inverse_transform(pca_images)
```

```python
def side_by_side(indexes):
    org = images[indexes].reshape(100, 100, 3)
    rec = np.rint(inv_pca_images[indexes].reshape(100, 100, 3)).astype(int)
    rec[rec<0] = 0
    rec[rec>255] = 255
    pair = np.concatenate((org, rec), axis=1)
    plt.figure(figsize=(10,5))
    plt.imshow(pair)
    plt.show()
    
for index in range(0,10):
    side_by_side(index)
```
